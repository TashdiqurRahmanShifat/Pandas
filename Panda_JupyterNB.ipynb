Find out the runs of V Kohli against opponent team

```python
runs=deliveries.groupby('batsman')
```

```python
vkruns=runs.get_group('V Kohli')
vkruns
```

```python
opponent=vkruns.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
opponent
```

```python
def scored_runs(batsman_name):
    all_batsman=deliveries.groupby('batsman')
    individual_batsman=all_batsman.get_group(batsman_name)
    return individual_batsman.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).iloc[0]
```

```python
scored_runs('V Kohli')
```

Find the most destructive batsman of IPL in (16-20) overs facing min 200 balls

```python
death_over_filter=deliveries['over']>15    # Fitlering
```

```python
death_over=deliveries[death_over_filter]   # DataFrame
```

```python
death_over
```

```python
all_batsman=death_over.groupby('batsman')   # All batsman 
```

```python
balls_Played=all_batsman.size()   # Balls faced in total
```

```python
balls_Played
```

```python
mask=balls_Played>200
#requiredPlayers=balls_Played
```

```python
requiredPlayer=balls_Played[mask]
```

```python
requiredPlayer
```

```python
requiredPlayerList=requiredPlayer.index.tolist()
```

```python
requiredPlayerList
```

```python
mask=death_over['batsman'].isin(requiredPlayerList)
```

```python
newPlayerDataframe=death_over[mask]
```

```python
newPlayerDataframe
```

```python
runs=newPlayerDataframe.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=newPlayerDataframe.groupby('batsman')['batsman_runs'].count()
```

```python
strike_rate=(runs/balls)*100
```

```python
strike_rate.sort_values(ascending=False).head(1)
```
